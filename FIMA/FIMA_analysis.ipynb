```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df1 = pd.read_csv("FimaNfipClaimsV1_1.csv", index_col = 0)
df2 = pd.read_csv("FimaNfipClaimsV1_2.csv", index_col = 0)
df3 = pd.read_csv("FimaNfipClaimsV1_3.csv", index_col = 0)
df = pd.concat([df1, df2, df3], ignore_index = True)

df
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
corr_matrix = df.select_dtypes(include = numerics).corr()

corr_matrix
```

```python
c = df.select_dtypes(include = numerics).corr().abs()
s = c.unstack()
so = s.sort_values(kind = "quicksort", ascending = False)

so
```
